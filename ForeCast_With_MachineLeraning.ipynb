##1-İMPORTİNG LİBRARİES

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

keras = tf.keras
```

##2-CREATE SERİES

```python
def plot_series(time, series, format="-", start=0, end=None, label=None):
    plt.plot(time[start:end], series[start:end], format, label=label)
    plt.xlabel("Time")
    plt.ylabel("Value")
    if label:
        plt.legend(fontsize=14)
    plt.grid(True)
    
def trend(time, slope=0):
    return slope * time
  
  
def seasonal_pattern(season_time):
    """Just an arbitrary pattern, you can change it if you wish"""
    return np.where(season_time < 0.4,
                    np.cos(season_time * 2 * np.pi),
                    1 / np.exp(3 * season_time))

  
def seasonality(time, period, amplitude=1, phase=0):
    """Repeats the same pattern at each period"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)
  
  
def white_noise(time, noise_level=1, seed=None):
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level
```

```python
time = np.arange(4 * 365 + 1)

slope = 0.05
baseline = 10
amplitude = 40
series = baseline + trend(time, slope) + seasonality(time, period=365, amplitude=amplitude)

noise_level = 5
noise = white_noise(time, noise_level, seed=42)

series += noise

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

**EVET SERİMİZİ OLUŞTURDUK ŞİMDİ TAHMİN KISMINA GEÇELİM**


##FORECASTİNG WİTH MACHİNE LEARNİNG

```python
def window_dataset(series,window_size,batch_size=32,shuffle_buffer =1000):
   dataset = tf.data.Dataset.from_tensor_slices(series)
   dataset = dataset.window(window_size +1,shift = 1,drop_remainder = True)
   dataset = dataset.flat_map(lambda window : window.batch(batch_size+1))
   dataset = dataset.shuffle(shuffle_buffer)
   dataset = dataset.map(lambda window : (window[:-1],window[-1]))
   dataset = dataset.batch(batch_size).prefetch(1)
   return dataset
```

```python
## TEST - VALİDASYON VERİLERİMİZİ BELİRLEYELİM

split_time = 1000

time_train = time[:split_time]
x_train = series[: split_time]
time_valid = time[split_time:]
x_valid = series[split_time :]
```

##LİNEAR MODEL

```python
keras.backend.clear_session()  # Bütün oturumları temizler
tf.random.set_seed(42)  # seed 'leri sabitleme ---->>> Kodu her çalıştırdığımızda aynı değeri vermek için
np.random.seed(42)

window_size = 30   
train_set = window_dataset(x_train,window_size)
valid_set = window_dataset(x_valid,window_size)

model = keras.models.Sequential([
    keras.layers.Dense(1,input_shape = [window_size]) # Giriş penceresini alıp o penceredeki değerlerin lineer kombinasyonlarını hesaplar ,tek nöronla
])

optimizer = keras.optimizers.SGD(lr = 1e-5,momentum = 0.9)  # stokastik gradyan iniş yöntemi kullanılır.(1'den - 5 kadar)

model.compile(loss = keras.losses.Huber(),  # Kayıpları huber yoluyla hesaplar
              optimizer = optimizer,
              metrics = ["mae"])

model.fit(train_set,epochs = 100,validation_data = valid_set)
```

**FENA DEĞİL DAHA İYİ DEĞERLER BULMAK İÇİN EL YORDAMIYLA GRADİENT İNİŞLERİ DEĞİŞTİREBİLİRİZ YADA BUNU OTOMATİK HALE GETİREBİLİRİZ**

##ÖĞRENME ORANI PROGRAMLAYICI YARDIMIYLA MODEL FİTTİNG

- Durum şu öğrenme oranı programlayıcı yardımıyla işlem esnasında öğrenme oranını değiştirebeliriz ve optimum değerlerde öğrenmesini sağlayabiliriz

```python
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)


window_size = 30

train_set = window_dataset(x_train,window_size)

model = keras.models.Sequential([
    keras.layers.Dense(1,input_shape = [window_size])
])

lr_schedule = keras.callbacks.LearningRateScheduler(  # 10 üzeri - 6 öğrenme oranı ile başla---> her 30 epoch'da 10 ile çarp 10^-6,10^-5,10^-4 ...
    lambda epoch : 1e-6 * 10**(epoch/30)  
)

optimizer = keras.optimizers.SGD(lr = 1e-6,momentum = 0.9)

model.compile(
    loss = keras.losses.Huber(),
    optimizer = optimizer,
    metrics = ["mae"]
)

history = model.fit(train_set,epochs = 100,callbacks = [lr_schedule])
```

**KADEMELİ OLARAK BİR YERE KADAR DÜŞTÜĞÜNÜ DAHA SONRA ARTTIĞINI GÖRÜYORUZ  MAE VE LOSS DEĞERLERİNİN BUNU GÖRSELLEŞTİRELİM**

```python
plt.figure(figsize = (10,6))
plt.semilogx(history.history["lr"],history.history["loss"])
plt.axis([1e-6,1e-3,0,20])
```

**BURADAKİ YORUM ŞU MODELİMİZ 10^-6 İLE 10^-4 ARASINDA KADEMELİ OLARAK ÖĞRENMESİNİ İYİLEŞTİRMİŞ VE MAE DEĞERİ KÜÇÜLMÜŞTÜR.SONRASINDA HIZLI BİR ŞEKİLDE ARTMIŞ BUNUN SEBEBİ AŞIRI ÖĞRENMEDEN O ZAMAN AŞIRI ÖĞRENME OLMADAN ÖNCE BİZ BU İŞLEMİ DURDURACAĞIZ VE EN OPTİMUM ÖĞRENME ORANINI ALACAZ**

**İŞLEME BAŞLAMADAN ÖNCE MODELİMİZ İÇİN 10^-5 OPTİMUM GİBİ DURUYOR O YÜZDEN SGD İNİŞ DEĞERİMİZ 1e-5 seçmek uygun gibi duruyor.

**PATLAMA YAŞANMADAN ÖNCE DURDURMAK İÇİN ERKEN DURDURMA İŞLEMİ YAPACAZ**

```python
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

# penceremizin aralığını belirleyelim
window_size = 30

# eğitim ve doğrulama setlerimizi belirleyelim
train_set = window_dataset(x_train,window_size)
valid_set = window_dataset(x_valid,window_size)

# modelimizi yaratalım(1(output) giriş shape = window_size kadar)
model = keras.models.Sequential([
    keras.layers.Dense(1,input_shape = [window_size]) # Tek katmanlı 1 outputlu
])

# optimizer ayarlayalım (1e-5)
optimizer = keras.optimizers.SGD(lr = 1e-5,momentum = 0.9)

# modelimizi compile yani derleyelim (loss yöntemi,optimizer,validason yöntemi)
model.compile(
    loss = keras.losses.Huber(),
    optimizer = optimizer,
    metrics = ["mae"]
)

## ERKEN DURDURUCU
early_stopping = keras.callbacks.EarlyStopping(patience = 10) # 10 epoch boyunca aşağı inme durursa durdurmayı çağır demek(patience)

# Ve modelimizi eğitmeye başlayalım
model.fit(
    train_set,
    epochs = 500,
    validation_data = valid_set,
    callbacks = [early_stopping]
)


```

**EVET MODELİMİZİ EĞİTTİK ŞİMDİ TAHMİN FONKSİYONU YAZALIM**

```python
def model_forecast(model,series,window_size):
  ds = tf.data.Dataset.from_tensor_slices(series)
  ds = ds.window(window_size ,shift =1,drop_remainder = True)
  ds = ds.flat_map(lambda w : w.batch(window_size))
  ds = ds.batch(32).prefetch(1)
  forecast = model.predict(ds)
  return forecast
```

```python
linear_forecast = model_forecast(model,series[split_time-window_size : -1],window_size)[:,0] # !!!   -----  [:,0] 2.boyuttan kurtulmak için(model 2boyutlu tahmin oluşturur diğer boyut 1 dir ondan kurtuluruz)
```

```python
linear_forecast.shape
```

**TAHMİN İLE GERÇEĞİ GÖRSELLEŞTİRELİM**

```python
plt.figure(figsize=(10,6))
plot_series(time_valid,x_valid,label = "Series")
plot_series(time_valid,linear_forecast,label = "Linear ForeCast")
plt.show()
```

```python
# ORTALAMA MUTLAK HATAYA BAKALIM 
keras.metrics.mean_absolute_error(x_valid,linear_forecast).numpy()
```

## DENSE MODEL FORECASTİNG

- Şİmdi iki katmanlı bir sinir ağı modeli kuralım

- Önce en iyi öğrenme modelini bulmamız gerekiyor

- En iyi öğrenme modelinde LearningRateSchuduler kullanacaz  !!!





```python
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

# window_size belirleyelim
window_size =30

# train setimizi belirleyelim
train_set = window_dataset(x_train,window_size)

# modelimizi yaratalım (10 nöronlu iki gizli katmanlı ve son olarak 1 nöronlu çıktı katmanı oluşturacaz --- relu = değer etkinleştirme işlemi)
model = keras.models.Sequential([
    keras.layers.Dense(10,activation = "relu",input_shape = [window_size]),
    keras.layers.Dense(10,activation = "relu"),
    keras.layers.Dense(1)
])

# 10^-7 den başlayarak her 20 epochda 10 ile çarpılacak bir learn callback oluşturuyoruz
lr_schedule = keras.callbacks.LearningRateScheduler(
    lambda epoch : 1e-7 * 10**(epoch/20)
)

# optimizer oluşturalım
optimizer = keras.optimizers.SGD(1e-7,momentum = 0.9)

# Compile yapalım 
model.compile(
    loss = keras.losses.Huber(),
    optimizer = optimizer,
    metrics = ["mae"]
)

history = model.fit(train_set,epochs = 100,callbacks = [lr_schedule])
```

**BELİRLİ BİR DEĞERDEN SONRA AŞIRI ÖĞRENME OLUP PATLADI BAKALIM NERDE PATLAMA YAŞANMIŞ**

```python
plt.figure(figsize = (10,6))
plt.semilogx(history.history["lr"],history.history["loss"])
plt.axis([1e-7,5e-3,0,30])
```

**GÖRÜNTÜYE GÖRE 1e-5 CİVARINDA TİTREME BAŞLADIĞI İÇİN OPTİMUM SEVİYE ORADAYMIŞ GİBİ ONU SEÇMEMİZ DAHA DOĞRU OLACAKTIR**

**İNİŞİMİZİ BULDUK ŞİMDİ OPTİMUM İNİŞ DEĞERLİ VE PATLAMA YAŞANMADAN ÖNCE DURMASINI SAĞLAYACAK MODELİMİZİ KURALIM**
- DURDURMA YAPACAĞIMIZ İÇİN **EarlyStopping** KULLANACAZ

```python
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)


# window - size belirleyelim
window_size = 30 

# train - valid setlerimizi belirleyelim 
train_set = window_dataset(x_train,window_size)
valid_set = window_dataset(x_valid,window_size)

# modelimizi kuralım (10 nöronlu 2 gizli katman ve 1 çıkış katmanından oluşacak -- activation = relu)
model = keras.models.Sequential([
    keras.layers.Dense(10,activation = "relu",input_shape = [window_size]),
    keras.layers.Dense(10,activation = "relu"),
    keras.layers.Dense(1)
])

# optimizer belirleyelim 
optimizer = keras.optimizers.SGD(lr = 1e-5,momentum = 0.9)

# Compile işlemi yapalım 
model.compile(
    loss = keras.losses.Huber(),
    optimizer = optimizer,
    metrics = ["mae"]
)

# Early Stopping 
early_stopping = keras.callbacks.EarlyStopping(patience = 10)

## MODEL EĞİTİMİNE BAŞLAYALIM
model.fit(
    train_set,
    epochs = 500,
    validation_data = valid_set,
    callbacks = [early_stopping]
)
```

**EĞİTİM BİTTİ ŞİMDİ EĞİTTİĞİMİZ MODEL İLE TAHMİN YAPALIM**

```python
dense_forecast = model_forecast(
    model,
    series[split_time-window_size : -1],
    window_size
)[:,0]
```

```python
plt.figure(figsize=(10,6))
plot_series(time_valid,x_valid,label = "Series Original")
plot_series(time_valid,dense_forecast,label ="Dense Forecast")
plt.show()
```

**ORTALAMA MUTLAK HATAYA BAKALIM**

```python
keras.metrics.mean_absolute_error(x_valid,dense_forecast).numpy()
```
